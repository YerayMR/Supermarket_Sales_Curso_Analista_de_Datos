```python
import os
import pandas as pd
import kagglehub

# Descargar dataset
path = kagglehub.dataset_download("faresashraf1001/supermarket-sales")
print("Carpeta del dataset:", path)

# Listar el contenido de la carpeta descargada para verificar el nombre del archivo
print("Contenido de la carpeta del dataset:")
print(os.listdir(path))

# Ruta al CSV principal
# Una vez que identifiques el nombre correcto del archivo, actualiza esta l√≠nea
csv_path = os.path.join(path, "SuperMarket Analysis.csv")

# Cargar datos tal cual
df = pd.read_csv(csv_path)
df.head()
```

```python
# Convertir columna Date a datetime (formato americano)
df["Date"] = pd.to_datetime(df["Date"], format="%m/%d/%Y")
```

```python
out_csv = "/content/supermarket_sales_clean_es.csv"
df.to_csv(out_csv, sep=";", decimal=",", index=False, encoding="utf-8-sig")
print("Guardado en:", out_csv)
```
